```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
cameraman = plt.imread("cameraman.tif")
plt.imshow(cameraman, cmap="gray");
```

```python
cameraman.shape, cameraman.dtype
```

```python
cameraman3x = np.stack([cameraman]*3, axis=-1)
cameraman3x.shape, cameraman3x.dtype
```

```python
np.array_equal(cameraman, cameraman3x[...,0])
```

```python
plt.imshow(cameraman3x);
```

### Conclusion
Visually, **stacking** 3-fold a grayscale image **looks no different** from the grayscale image itself.
